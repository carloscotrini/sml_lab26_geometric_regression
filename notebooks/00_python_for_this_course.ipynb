## 1. Numpy: Arrays and Operations

Everything in this course is a vector or a matrix. Numpy makes them easy.

````python
# Creating arrays
heights = np.array([165, 170, 175, 180, 185, 190, 172, 168, 177, 183])  # cm
weights = np.array([62, 68, 73, 80, 85, 92, 70, 65, 76, 82])            # kg

print(f"Heights: {heights}")
print(f"Shape:   {heights.shape}")  # (10,) means 10 elements
print(f"Zeros:   {np.zeros(5)}")
print(f"Random:  {np.random.randn(5).round(2)}")
````

````python
# Array operations — pay special attention to @
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

print(f"a + b  = {a + b}")        # element-wise addition
print(f"a * b  = {a * b}")        # element-wise multiply
print(f"a @ b  = {a @ b}")        # dot product — THIS ONE MATTERS MOST

# Matrix multiply with @
M = np.array([[1, 2], [3, 4], [5, 6]])  # shape (3, 2)
v = np.array([10, 20])                   # shape (2,)
print(f"M @ v  = {M @ v}")        # matrix-vector product
print(f"M.T shape = {M.T.shape}")  # transpose: (3,2) -> (2,3)
````

````python
# Indexing
print(f"First height:    {heights[0]}")     # single element
print(f"Last 3 heights:  {heights[-3:]}")   # slicing from the end

# For matrices: [row, column]
print(f"M first row:     {M[0]}")           # [1, 2]
print(f"M first column:  {M[:, 0]}")        # [1, 3, 5]
````

## 2. Matplotlib: Seeing Your Data

Every insight in this course arrives as a picture. Here's how to make them.

````python
# Basic scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(heights, weights, color="#3B82F6", s=60)
plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("Height vs Weight")
plt.tight_layout()
plt.show()
````

````python
# Subplots: two plots side by side
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.scatter(heights, weights, color="#3B82F6", s=60)
ax1.set_title("Scatter")
ax1.set_xlabel("Height (cm)")

ax2.hist(weights, bins=6, color="#10B981", edgecolor="white")
ax2.set_title("Distribution of Weights")
ax2.set_xlabel("Weight (kg)")

plt.tight_layout()
plt.show()
````

````python
# 3D plotting — you will see a lot of this in the course
fig = plt.figure(figsize=(7, 5))
ax = fig.add_subplot(111, projection="3d")

z = heights + weights + np.random.randn(10) * 5
ax.scatter3D(heights, weights, z, color="#F59E0B", s=60)
ax.set_xlabel("Height")
ax.set_ylabel("Weight")
ax.set_zlabel("Z")
ax.set_title("3D Scatter")
plt.tight_layout()
plt.show()
````

## 3. Statsmodels: Running a Regression

Five commands. That's all you need.

````python
# Fit an OLS regression: weight ~ height
X = sm.add_constant(heights)       # adds a column of 1s for the intercept
model = sm.OLS(weights, X).fit()   # OLS(y, X) — note: y comes first

print(model.summary())
````

````python
# The four results you will use constantly
print(f"Coefficients:  {model.params.round(3)}")              # intercept and slope
print(f"Fitted values: {model.fittedvalues[:3].round(1)}...")  # predicted y
print(f"Residuals:     {model.resid[:3].round(2)}...")         # errors
print(f"R-squared:     {model.rsquared:.3f}")                 # how good the fit is
````

````python
# One more: the hat matrix diagonal (you will learn what this means later)
from statsmodels.stats.outliers_influence import OLSInfluence

influence = OLSInfluence(model)
leverage = influence.hat_matrix_diag
print(f"Leverage values: {leverage.round(3)}")
````

## 4. Interactive Widgets

Many cells in this course have sliders, draggable points, or toggles. Drag a slider and the plot updates instantly.

````python
# Interactive slider demo — drag to change the slope
try:
    from ipywidgets import interact, FloatSlider

    def plot_line(slope=1.0):
        plt.figure(figsize=(6, 4))
        plt.scatter(heights, weights, color="#3B82F6", s=60)
        pred = slope * (heights - heights.mean()) + weights.mean()
        plt.plot(heights, pred, color="#EF4444", linewidth=2)
        plt.title(f"Slope = {slope:.1f}")
        plt.xlabel("Height (cm)")
        plt.ylabel("Weight (kg)")
        plt.tight_layout()
        plt.show()

    interact(plot_line, slope=FloatSlider(min=0, max=3, step=0.1, value=1.0))
except ImportError:
    print("Widgets not available. Set INTERACTIVE = False in the course notebooks.")
````
